## Bar Chart
- Used to respresent a comparison between categories of data.

```python
# Up Size
plt.figure(figsize=(10,6))

# Count Plot
sns.countplot(x='Type 1',data=pokemon)
```

### Order the Bar Chart

```python
pokemon["Type 1"].value_counts()
```

```python
pokemon["Type 1"].value_counts().index
```

```python
# Up Size
plt.figure(figsize=(14,6))

# Count Plot
sns.countplot(x='Type 1',
              data=pokemon,
              order = pokemon['Type 1'].value_counts().index,
              color = '#56CCF2')
```

### Can we Adjust more X-Axis label for better readble

```python
# Up Size
plt.figure(figsize=(14,6))
plt.title("Count Pokemon Type 1")

#Count Plot
sns.countplot(x='Type 1',
              data=pokemon,
              palette='rainbow',
              order = pokemon['Type 1'].value_counts().index)

plt.xticks(rotation=90)


# rcParams method of matplotlib to increase the size of x-axis values and labels of x-axis.
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['axes.labelsize'] = 20
plt.rc('axes', titlesize = 30)
```

## Box Chart
- Used to represent the statistical five number summary of a data set i.e. Minimum, First Quartile, Median, Third Quartile, and Maximum.

- Used for exploratory data analysis. Through this, we can represent the shape og the distribution of data.

```python
# Up Size
plt.figure(figsize=(14,6))


#Boxplot
sns.boxplot(data=pokemon)
plt.xticks(rotation=90)
```

```python
# Try drop 3 columns of distinguished
df_drop = pokemon.drop(columns=['Total','Generation','Legendary'])

# Up Size
plt.figure(figsize=(14,6))
sns.boxplot(data=df_drop)
```

## Violin Chart
- Help for representing a fantastically huge amount of information effectively in the small amount of space.

```python
# Setting the theme
sns.set_style('darkgrid')
plt.figure(figsize=(14,6))

# Violin Plot
vio_plot = sns.violinplot(x='Type 1', y='Attack', data=pokemon)

plt.xticks(rotation=30)
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams["axes.labelsize"] = 15
```

## Swarm Chart
-  This chart is better representation of the distribution of values. 

- It can be built on its own but is also a good complement to a box or violin plot.

```python
plt.figure(figsize=(14,6))
plt.xticks(rotation=30)

sns.swarmplot(x='Type 1', y='Attack', data=pokemon, dodge=False)

plt.rcParams['xtick.labelsize'] = 15
plt.rcParams["axes.labelsize"] = 15
```

### Overlaying Swarm and Violin Plots

- Overlaying help us to analyse in a much more efficient manner.

- Helps for visualizing the distribution of data and the probability density.

```python
#Setting the figure with matplotlib
plt.figure(figsize=(14,6))
plt.xticks(rotation=30)
plt.rcParams["axes.labelsize"] = 15

#Creating the desired plot
sns.violinplot(x='Type 1',y='Attack',data=pokemon,
              inner=None #removes the inner bars inside the violins
              )

sns.swarmplot(x='Type 1',y='Attack',data=pokemon)

#Title for the plot
plt.title('Attacks by various types')
```

```python
#Setting the figure with matplotlib
plt.figure(figsize=(10,6))
plt.xticks(rotation=30)
plt.rcParams["axes.labelsize"] = 15

#Creating the desired plot
sns.violinplot(x='Type 1',y='Attack',data=pokemon,
              inner=None #removes the inner bars inside the violins
              )

# We can make distributed of plot more clear by using "marker" Black color
sns.swarmplot(x='Type 1',y='Attack',data=pokemon,
             color='k',#for making the points black
             alpha=0.6) #value of alpha will increase the transparency

#Title for the plot
plt.title('Attacks by various types')
```

## Joint Kernel Density Plot

- Help for finding the probability density function of any dataset.

- Help in smoothing the data around values of PDF (Probability density function).

```python
sns.set_style('whitegrid')

plt.figure(figsize=(14,6))
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15

# sns.kdeplot(pokemon.Attack, pokemon.Defense)
sns.kdeplot(pokemon['Attack'], shade=True)
```

## Joint Plot
- Get the liberty to use two plots for representation of the same data which helps in a better analysis.

```python
plt.figure(figsize=(14,6))
sns.jointplot(x="Sp. Atk",y='Sp. Def',data=pokemon,
              kind='scatter',color='#cbb4d4')
```

## Heatmap

- It's the 2-dimensional figure

- Helps for visualizing complex data. We can even use it to represent large datasets.

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

### Displaying the flights dataset.

```python
flights.T.head()
```

```python
# Create Pivot Table
flights_pivot = flights.pivot(index='month',columns='year',values='passengers')
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(flights_pivot,cmap="OrRd")
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(flights_pivot,cmap="coolwarm",annot=True,fmt='d',linewidths=2)
plt.xticks(rotation=70)
```
